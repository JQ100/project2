```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/superstore - superstore.csv')
df.head()
```

```python
# address duplicate rows
df[df.duplicated(keep=False)]
# there are no duplicates :)
```

```python
# delete unnecessary columns
df = df.drop(['Row ID', 'Ship Date', 'Ship Mode', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region', 'Quantity'], axis=1)
df.head()
# we only need to preserve order information, no need shipment info that only the customer will want
```

```python
# address missing values
print(df.isna().sum().sum(), 'missing values')
# there are no missing values :)
```

```python
# Visual 1
feature1 = df['Category'].values
feature2 = df['Sales'].values
plt.bar(feature1, feature2)
plt.legend()
```

According to the bar chart, we have the most sales in the technology sector and the least sales in the furniture sector.

```python
# Visual 2
feature3 = df['Profit'].values
plt.scatter(feature1, feature3)
```

Based on the scatter chart visual above, we notice a greater range of profit/loss in the technology sector. We also notice major outliers here. And combined with the first visual, we can conclude that greater sales contribute to a wider varying profit/loss.
